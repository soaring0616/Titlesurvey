```python
import pandas as pd
fname = 'church/03 biblemining/titleofmattandeph.csv'
df = pd.read_csv(fname)
df.head()
```

```python
df
```

```python
# -- 斷句 -- #
# https://ithelp.ithome.com.tw/articles/10192043 

# -- Setting -- #
import jieba
jieba.set_dictionary('dictionary/dict.txt') 
with open('dictionary/stops.txt', 'r', encoding='utf8') as f:
    stops = f.read().split('\n') 
    
stops.append('\n')
stops.append('\n\n')
stops.append('）。')
stops.append('），')
stops.append('-')
stops.append(':')
stops.append('：')
stops.append('－')
stops.append(':"')
stops.append('─')
stops.append('廿')
stops.append('）、')
stops.append('"')
stops.append(' ')
stops.append('20')
stops.append('10')


# -- Cutting Matthew -- #
from collections import Counter


mattls = df['馬太福音'].tolist()
mattgenre = df['種類'].tolist()

mattlscut = []
matttop5 = []

for textStr in mattls:
    terms = [t for t in jieba.cut(textStr, cut_all=True) if t not in stops]
    dic = {}
    top5=(sorted(Counter(terms).items(), key=lambda x:x[1], reverse=True)[:5])
    matttop5.append([item for item, _ in top5])
    for kw, fre in sorted(Counter(terms).items(), key=lambda x:x[1], reverse=True):
        dic[kw] = fre
    mattlscut.append(dic)


dfmattls = pd.DataFrame(mattls, columns=['內文'])
dfmattls['斷詞'] = mattlscut
dfmattls.head()


txtlenls=[]
for item in mattls:
    txtlenls.append(len(item))

dfmattls = pd.DataFrame(mattls, columns=['內文'])
dfmattls['名稱'] = df['名稱'].tolist()
dfmattls['內文長度'] = txtlenls
dfmattls['種類'] = mattgenre
dfmattls['5 keywords'] = matttop5
dfmattls['斷詞'] = mattlscut
```

```python
dfmattls
```

```python
exls = []
for item in matttop5:
    for x in item:
        exls.append(x)

top5=(sorted(Counter(exls).items(), key=lambda x:x[1], reverse=True)[:5])
print(top5)
```
